```python
import pandas as pd

#CHANGE HERE FOR THE CORRECT PATH OF OUTPUT.CSV
data = pd.read_csv(r"C:\Users\HP\Downloads\databases group project github materials\q5_code\output.csv")

#Inspect the data
print(data.head())
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
#Scatter plot
plt.figure(figsize=(10, 6))
sns.scatterplot(data=data, x='author_count', y='popularity_score', s=100, color='blue')

plt.title('Number of Authors vs Popularity Score', fontsize=16)
plt.xlabel('Number of Authors', fontsize=12)
plt.ylabel('Popularity Score', fontsize=12)
plt.grid(True)
plt.show()
```

```python
#Group by author count and calculate average popularity
grouped_data = data.groupby('author_count')['popularity_score'].mean().reset_index()

#Bar chart
plt.figure(figsize=(10, 6))
sns.barplot(data=grouped_data, x='author_count', y='popularity_score', palette='viridis')

plt.title('Average Popularity Score by Number of Authors', fontsize=16)
plt.xlabel('Number of Authors', fontsize=12)
plt.ylabel('Average Popularity Score', fontsize=12)
plt.xticks(rotation=45)
plt.show()
```

```python
#Heatmap
heatmap_data = pd.pivot_table(data, values='popularity_score', index='author_count', aggfunc=np.mean)

plt.figure(figsize=(8, 5))
sns.heatmap(heatmap_data, annot=True, cmap='coolwarm', fmt=".2f")

plt.title('Heatmap of Popularity Score by Author Count', fontsize=16)
plt.xlabel('Popularity Score', fontsize=12)
plt.ylabel('Number of Authors', fontsize=12)
plt.show()
```

```python
#Correlation analysis
correlation = data[['author_count', 'popularity_score']].corr()
print("Correlation Matrix:\n", correlation)

sns.lmplot(data=data, x='author_count', y='popularity_score', height=6, aspect=1.5)
plt.title('Regression: Authors vs Popularity', fontsize=16)
plt.xlabel('Number of Authors', fontsize=12)
plt.ylabel('Popularity Score', fontsize=12)
plt.show()
```
